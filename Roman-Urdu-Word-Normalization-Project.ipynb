Roman Urdu lacks standard lexicon and usually many spelling variations exist for a given word,
e.g., the word zindagi (life) is also written as zindagee, zindagy, zaindagee and zndagi.
So, in this question you have to Normalize Roman Urdu words using the following Rules given in the attached Pdf.
Your Code works for a complete Sentence or multiple sentences.

For Example: zaroori, zaruri, zarori map to the 'zrory'. 
So zrory becomes the correct word for all representations mentioned above.

```python
#Your code here
import re
# Taking input from the user
user_input = input("Enter a string: ")

# Printing the input string
##print("You entered:", user_input)

# Check if the string ends with "ain"
if user_input.endswith("ain"):
    user_input=user_input.replace("ain","ein")    # Replace "ain" with "ein"
##print(user_input)
modified_input = re.sub(r'(?<!^)ar', 'r', user_input) ## Replace 'ar' except at the start with 'r'

user_input = re.sub(r'(iy|y+)', 'i', user_input)    ## Replace 'iy' and multiple 'y' with 'i'

if user_input.endswith("ay"):      ## Replace 'ay' at the end with 'e'
    user_input=user_input.replace("ay","e")


user_input = re.sub(r'(ih|h+)', 'eh', user_input)   ## replace 'ih' and multiple 'h' with 'eh'



if user_input.endswith("ey"):    ## Replace 'ey' at the end with 'e'
    user_input=user_input.replace("ey","e")
    
user_input = re.sub(r'(s+)', 's', user_input)    ## Replace multiple 's' with 's'  


if user_input.endswith("ie"):    ## Replace 'ie' at the end with 'y'
    user_input=user_input.replace("ie","y")

    

    
    user_input = re.sub(r'ry(?!$)', 'ri', user_input)     ### Replace 'sy' except at the end with 'si'

    
if  user_input.startswith("es"):     ## Replace 'es' at the start with 'is'
    user_input=user_input.replace("es","is")
    

user_input = re.sub(r'sy(?!$)', 'si', user_input)     ## Replace 'sy' except at the end with 'si'


user_input = re.sub(r'(a+)', 'a', user_input)     ## Multiple 'a' with 'a'

user_input = re.sub(r'ty(?!$)', 'ti', user_input)    ## Replace 'ty' except at the end with 'ti'

user_input = re.sub(r'(j+)', 'j', user_input)    ## Multiple 'j' with 'j'

user_input = re.sub(r'(o+)', 'o', user_input)    ## Multiple 'o' with 'o'

user_input = re.sub(r'(ee+)', 'i', user_input)  ## Multiple 'ee' with 'i'

user_input = re.sub(r'(d+)', 'd', user_input)   ## Multiple 'd' with 'd'

user_input = re.sub(r'(u+)', 'o', user_input)    ## Multiple 'u' with 'o'

if user_input.endswith("i"):    ## Replace 'i' at the end with 'y'
    user_input=user_input.replace("i","y")
    
if  user_input.startswith("za"):     ## Replace 'za' at the start with 'z'
    user_input=user_input.replace("za","z")

print(user_input)
```
